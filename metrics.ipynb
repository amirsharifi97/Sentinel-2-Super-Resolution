```python
from skimage.metrics import peak_signal_noise_ratio, structural_similarity
def calculate_metrics_train(model, train_img, train_label):
    model.eval()
    total_psnr = 0.0
    total_ssim = 0.0

    with torch.no_grad():
        for i in range(0, len(train_img), batch_size):
            batch_img = train_img[i:i + batch_size].to(device)
            batch_label = train_label[i:i + batch_size].to(device)

            output = model(batch_img)
            output = output.clamp(0, 1)

            batch_img = batch_img.cpu().numpy().transpose(0, 2, 3, 1)
            output = output.cpu().numpy().transpose(0, 2, 3, 1)
            batch_label = batch_label.cpu().numpy().transpose(0, 2, 3, 1)

            for j in range(len(batch_img)):
                psnr = peak_signal_noise_ratio(batch_label[j], output[j])
                ssim = structural_similarity(batch_label[j], output[j], multichannel=True)
                total_psnr += psnr
                total_ssim += ssim

    avg_psnr = total_psnr / len(train_img)
    avg_ssim = total_ssim / len(train_img)

    return avg_psnr, avg_ssim

# Move the model and training data to the same device
model = model.to(device)
train_img = train_img.to(device)
train_label = train_label.to(device)

# Calculate PSNR and SSIM for training set
train_psnr, train_ssim = calculate_metrics_train(model, train_img, train_label)
print(f"Train PSNR: {train_psnr:.4f}")
print(f"Train SSIM: {train_ssim:.4f}")

```
